```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

print(os.listdir())

df = pd.read_csv("all_data.csv")
print(df)

# Convert Year to a categorical or numeric type
df['Year'] = df['Year'].astype(int)

# Summary Statistics
summary_stats = df.groupby('Country').agg({
    'Life expectancy at birth (years)': ['mean', 'median', 'min', 'max', 'std'],
    'GDP': ['mean', 'median', 'min', 'max', 'std']
})
display(summary_stats)

# Correlation between GDP and Life Expectancy
correlation = df[['GDP', 'Life expectancy at birth (years)']].corr()
display(correlation)

# Line Plot for Life Expectancy Over Time
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='Life expectancy at birth (years)', hue='Country')
plt.title('Life Expectancy Over Time')
plt.xlabel('Year')
plt.ylabel('Life Expectancy (Years)')
plt.legend(title='Country')
plt.grid(True)
plt.show()

# Line Plot for GDP Over Time
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='GDP', hue='Country')
plt.title('GDP Over Time')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.legend(title='Country')
plt.grid(True)
plt.show()

# Box Plot: Life Expectancy by Country
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Country', y='Life expectancy at birth (years)')
plt.title('Life Expectancy Distribution by Country')
plt.xlabel('Country')
plt.ylabel('Life Expectancy (Years)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Box Plot: GDP by Country
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Country', y='GDP')
plt.title('GDP Distribution by Country')
plt.xlabel('Country')
plt.ylabel('GDP')
plt.xticks(rotation=45)
plt.yscale('log')  # Log scale for better visualisation
plt.grid(True)
plt.show()

# Bar Chart: Average Life Expectancy by Country
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Country', y='Life expectancy at birth (years)', estimator=lambda x: x.mean())
plt.title('Average Life Expectancy by Country')
plt.xlabel('Country')
plt.ylabel('Average Life Expectancy (Years)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Bar Chart: Average GDP by Country
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Country', y='GDP', estimator=lambda x: x.mean())
plt.title('Average GDP by Country')
plt.xlabel('Country')
plt.ylabel('Average GDP')
plt.xticks(rotation=45)
plt.yscale('log')  # Log scale for better visualisation
plt.grid(True)
plt.show()
```
